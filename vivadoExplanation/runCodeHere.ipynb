```python
axi_addr=ol.ip_dict['dual_port_axi_instr_0']['phys_addr']
axi_range=ol.ip_dict['dual_port_axi_instr_0']['addr_range']
print(axi_addr,axi_range)#Slave segment '/custom_full_0/S00_AXI/S00_AXI_mem' is being assigned into address space '/processing_system7_0/Data' at <0x7AA0_0000 [ 64K ]>.
```

```python

from pynq import MMIO
mmio=MMIO(axi_addr,axi_range)
```

```python
read_off= 0x4
read_data= mmio.read(read_off)
print(bin(read_data))
print(hex(read_data))
```

```python
write_too = 0x0

bit_string = '0010010000000100'

# Step 2: Convert binary to integer
int_value = int(bit_string, 2)

# Step 3: Convert integer to hexadecimal
hex_value = hex(int_value)

# Print the hexadecimal value
print(hex_value)

mmio.write(write_too, int_value)

# Read back
val = mmio.read(write_too)
print(bin(val))  
print(hex(val))  
```

```python
#RESET 0X0 AND 0X4:
mmio.write(0x0, 0x0)
mmio.write(0x4, 0x0)
```

```python
for read_off in range(0x0, 0x10000, 0x1):  # Assuming 4-byte (32-bit) memory reads
    try:
        read_data = mmio.read(read_off)
        if read_data != 0:
            print(f"Address: {hex(read_off)} | Data: {bin(read_data)} | {hex(read_data)}")
    except:
        continue
```

```python
for i in range(256):
    mmio.write(4 * i, i)  # Write value `i` to memory slot `i`

# Read back the value at the first slot to confirm
value = mmio.read(0)  # Read from the first memory location
print("First slot value:", value)

# Optionally, read and print all the slots
for i in range(256):
    print(f"Slot {i} value:", mmio.read(4 * i))
```
